## 교차 검증

* K 폴드

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np

iris = load_iris()
features = iris.data
label= iris.target
dt_clf = DecisionTreeClassifier()

# KFold 폴드수는 5
kfold = KFold(n_splits=5)
cv_accuracy = []
print("데이터 세트 크기:", features.shape[0])
```

```python
n_iter = 0

for train_index, test_index in kfold.split(features):
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]
    dt_clf.fit(X_train, y_train)
    pred = dt_clf.predict(X_test)
    n_iter += 1
    # 정확도 측정
    accuracy = np.round(accuracy_score(y_test, pred), 4)
    train_size = X_train.shape[0]
    test_size = X_test.shape[0]
    print("\n#{0} 교차 검증 정확도: {1}, 학습 데이터 크기: {2}, 검증 데이터 크기: {3}"
          .format(n_iter, accuracy, train_size, test_size))
    print("# {0} 검증 세트 인덱스: {1}".format(n_iter, test_index))
    cv_accuracy.append(accuracy)

print("평균 검증 정확도:", np.mean(cv_accuracy))
```

### Stratified K 폴드
* 폴드마다 라벨별 데이터 분포도가 불균등한 것을 해결하기 위한 교차검증법
* 전체 데이터의 라벨 분포도를 고려하여 폴드를 나눈다.

```python
import pandas as pd

iris = load_iris()

iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['label']=iris.target
iris_df['label'].value_counts() # 라벨별로 1:1:1의 분포를 가짐을 확인할 수 있음.
```

```python
kfold = KFold(n_splits=3)
n_iter=0
for train_index, test_index in kfold.split(iris_df):
    n_iter += 1
    label_train = iris_df['label'].iloc[train_index]
    label_test = iris_df['label'].iloc[test_index]
    print("교차 검증: {0}".format(n_iter))
    print("학습 레이블 데이터 분포:", label_train.value_counts())
    print("검증 레이블 데이터 분포:", label_test.value_counts())
```

## 레이블 인코딩 (Label encoding)
* 라벨마다 숫자를 붙여서 인코딩하는 방식
* 숫자의 크기를 중요도로 인식한다는 문제점이 있음

```python
from sklearn.preprocessing import LabelEncoder

items = ['TV', '냉장고', '전자레인지', '컴퓨터', '선풍기', '선풍기', '믹서', '믹서']

encoder = LabelEncoder() # 객체 생성
encoder.fit(items)
labels = encoder.transform(items)
print('인코딩 변환값:', labels)
```

```python
# encoder의 classes_ 피처는 0부터 순서대로 원본 라벨값을 가진다.
print("인코딩 클래스:", encoder.classes_) 
```

```python
# inverse_transform은 디코딩을 수행한다.
print("디코딩 원본 값:", encoder.inverse_transform([4,5,2,0,1,1,3,3]))
```

## 원-핫 인코딩(One-Hot encoding)
* 라벨인코딩의 숫자 크기 문제를 해결하기 위한 인코딩 방식이다.
* 숫자가 아닌 넘파이 형태로 특정 컬럼에만 1부여, 나머지는 0 이런 식으로 한다.

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

items = ['TV', '냉장고', '전자레인지', '컴퓨터', '선풍기', '선풍기', '믹서', '믹서']

# 2차원 ndarray로 변환한다.
items = np.array(items).reshape(-1, 1)
# 원핫 인코딩
oh_encoder = OneHotEncoder() # 객체 생성
oh_encoder.fit(items)
oh_labels = oh_encoder.transform(items)
# 원핫 변환 결과는 희소행렬이라서 toarray()를 이용해 밀집 행렬로 변환
print(oh_labels.toarray())
print(oh_labels.shape)
```

## 피처 스케일링과 정규화

## StandardScaler
* 표준화를 수행하는 함수
* (원본값-평균)/표준편차
* 0~1 사이의 값을 반환

```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
iris_data = iris.data
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)

print("피처 평균값: ", iris_df.mean())
print("피처 분산값: ", iris_df.var())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)

iris_df_scaled = pd.DataFrame(data=iris_scaled, columns=iris.feature_names)
print("피처 평균값 ", iris_df_scaled.mean())
print('피처 분산값 ', iris_df_scaled.var())
```

### MinMaxSclaer
* 그냥 0~1 사이의 값으로 반환
* 음수가 있을 때는 -1~1

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)

iris_df_scaled = pd.DataFrame(data=iris_scaled, columns=iris.feature_names)
print("피처 최솟값: ", iris_df_scaled.min())
print('파처 최대값: ', iris_df_scaled.max())
# 결과가 0,1임을 확인할 수 있다
```
