# 정렬, Aggregation함수, GroupBy 적용

* DF, Series의 정렬 - sort_values()
* ()안에 by='칼럼명'을 입력하면 해당 칼럼을 기준으로 정렬해줌
* 기본값은 오름차순임.
* 오름차순: ascending=True
* 내림차순: ascending=False

```python
titanic_sorted = titanic_df.sort_values(by=['Name'])
titanic_sorted.head()
```

```python
titanic_sorted = titanic_df.sort_values(by=['Pclass', 'Name'], ascending=False)
titanic_sorted.head()
```

* Aggregation 함수 적용
* 연산을 수행하는 함수임.

```python
titanic_df.count()
```

* groupby(): 칼럼명 변수를 기준으로 그룹화함.

```python
titanic_groupby = titanic_df.groupby(by='Pclass')
print(type(titanic_groupby)) # df를 반환함
```

```python
titanic_groupby = titanic_df.groupby('Pclass').count()
titanic_groupby
```

```python
titanic_groupby = titanic_df.groupby('Pclass')[['PassengerId', 'Survived']].count()
titanic_groupby
```

```python
titanic_df.groupby('Pclass')['Age'].agg([max, min])
# 여러 종류의 연산을 수행하고 싶을 때는 agg() 안에 리스트의 형태로 연산자를 작성
```

```python
agg_format={'Age':'max', 'SibSp':'sum', 'Fare':'mean'}
titanic_df.groupby('Pclass').agg(agg_format)
```

# 결손 데이터 처리하기

## isna()로 결손 데이터 여부 확인
* 그냥 isna()를 입력하면 해당 칼럼이 "모두" Null 값인지 확인가능
* Null값의 개수를 확인하고 싶을 때는 isna().sum()

```python
titanic_df.isna().head(3)
```

```python
titanic_df.isna().sum()
```

## fillna()로 결손 데이터 대체하기

```python
titanic_df['Cabin']=titanic_df['Cabin'].fillna('C000')
titanic_df.head(3)
```

```python
titanic_df['Age']=titanic_df['Age'].fillna(titanic_df['Age'].mean()) # 평균값으로 대체함.
titanic_df['Embarked']=titanic_df['Embarked'].fillna('S')
titanic_df.isna().sum()
```
